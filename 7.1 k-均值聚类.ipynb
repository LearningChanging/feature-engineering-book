## 例7-1 生成 k -均值示例的代码

```python
import numpy as np
from sklearn.cluster import KMeans
from sklearn.datasets import make_blobs
import matplotlib.pyplot as plt
import seaborn as sns
# %matplotlib notebook
```

```python
n_data = 1000
seed = 1
n_centers = 4

# Generate random Gaussian blobs and run K-means
# 生成符合高斯随机分布的点团，并运行k-均值算法
blobs, blob_labels = make_blobs(n_samples=n_data, n_features=2,
                                centers=n_centers, random_state=seed)
clusters_blob = KMeans(n_clusters=n_centers,
                       random_state=seed).fit_predict(blobs)

# Generate data uniformly at random and run K-means
# 生成完全随机的数据，并运行k-均值算法
uniform = np.random.rand(n_data, 2)
clusters_uniform = KMeans(n_clusters=n_centers,
                          random_state=seed).fit_predict(uniform)
```

```python
# 对结果进行可视化的Matplotlib代码
figure = plt.figure(figsize=(16, 9))
plt.subplot(221)
plt.scatter(blobs[:, 0], blobs[:, 1], c=blob_labels,
            edgecolors='k', cmap='PuRd') # edgecolors 点的外边颜色
plt.title("(a) Four randomly generated blobs", fontsize=14)
plt.axis('off') # 去掉边框

plt.subplot(222)
plt.scatter(blobs[:, 0], blobs[:, 1], c=clusters_blob,
            edgecolors='k', cmap='PuRd')
plt.title("(b) Clusters found via K-means", fontsize=14)
plt.axis('off')

plt.subplot(223)
plt.scatter(uniform[:, 0], uniform[:, 1], edgecolors='k')
plt.title("(c) 1000 randomly generated pois", fontsize=14)
plt.axis('off')

plt.subplot(224)
plt.scatter(uniform[:, 0], uniform[:, 1],
            c=clusters_uniform, edgecolors='k', cmap='PuRd')
plt.title("(d) Clusters found via K-means", fontsize=14)
plt.axis('off')
```
