```python
import pandas as pd
import json
```

```python
def load_json_df(filename, num_bytes=-1):
    """"
    Load the first `num_bytes` of the filename as a json blob, 
    convert each line into a row in a Pandas data frame.
    """
    with open(filename) as fs:
        df = pd.DataFrame([json.loads(x) for x in fs.readlines(num_bytes)])
    return df
```

```python
biz_df = load_json_df('./data/yelp_dataset/business.json')
```
